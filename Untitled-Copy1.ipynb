```python
def printCompasDetails(data, *seeds, mask=()):
    """
    Function to print the full Compas output for given seeds, optionally with an additional mask
    """
    list_of_keys = list(data.keys())

    # Check if seed parameter exists - if not, just print without (e.g RunDetails)
    if ('SEED' in list_of_keys) | ('SEED>MT' in list_of_keys): # Most output files 
        #SEED>MT is a relic from older versions, but we leave this in for backwards compatibility

        # Set the seed name parameter, mask on seeds as needed, and set the index
        seedVariableName='SEED' if ('SEED' in list_of_keys) else 'SEED>MT'
        list_of_keys.remove(seedVariableName) # this is the index above, don't want to include it
    
        allSeeds = data[seedVariableName][()]
        seedsMask = np.in1d(allSeeds, seeds)
        if len(seeds) == 0: # if any seed is included, do not reset the mask
            seedsMask = np.ones_like(allSeeds).astype(bool)
        if mask == ():
            mask = np.ones_like(allSeeds).astype(bool)
        mask &= seedsMask

        df = pd.DataFrame.from_dict({param: data[param][()][mask] for param in list(data.keys())}).set_index(seedVariableName).T

    else: # No seed parameter, so do custom print for Run Details

        # Get just the keys without the -Derivation suffix - those will be a second column
        keys_not_derivations = []
        for key in list_of_keys:
            if '-Derivation' not in key:
                keys_not_derivations.append(key)
        
        # Some parameter values are string types, formatted as np.bytes_, need to convert back
        def convert_strings(param_array):
            if isinstance(param_array[0], np.bytes_):
                return param_array.astype(str)
            else:
                return param_array

        df_keys = pd.DataFrame.from_dict({param: convert_strings(data[param][()]) for param in keys_not_derivations }).T
        nCols = df_keys.shape[1] # Required only because if we combine RDs, we get many columns (should fix later)
        df_keys.columns = ['Parameter']*nCols
        df_drvs = pd.DataFrame.from_dict({param: convert_strings(data[param+'-Derivation'][()]) for param in keys_not_derivations }).T
        df_drvs.columns = ['Derivation']*nCols
        df = pd.concat([df_keys, df_drvs], axis=1)

    # Add units as first col
    units_dict = {key:data[key].attrs['units'].astype(str) for key in list_of_keys}
    df.insert(loc=0, column='(units)', value=pd.Series(units_dict))
    return df
```

```python
## goal: for each formed Compact Object (CO), determine the time it spent as a stripped star
#
## methodology
# 1) mask the different types of CO's --> SPs['Stellar_Type(1/2)'][()] == [10, 11, 12, 13, 14]
## for each type:
# 2) determine time spent as stripped star
```

```python
## further improvements
# 1) bound vs. unbound --> bound_mask = SPs['Unbound'][()] == 0
# 2) stripped in binary process or not --> binary_stripped_mask = (RLOF['Stellar_Type(1)>MT'][()] >= 7) & (RLOF['Stellar_Type(1)>MT'][()] <= 9)
```

```python
def MassVsTimeSpentStripped(CoType, BoundUnbound, binaryStripped):
    
    ## For the primary companion:
    
    # 1) mask for stellar type, bound vs unbound
    StellarType_mask1 = (SPs['Stellar_Type(1)'][()] == CoType) & (SPs['Unbound'][()] == BoundUnbound)
    StellarType_seeds1 = SPs['SEED'][StellarType_mask1]
    
    # 2) mask for binary stripped stars
    BinaryStripped_mask1 = (RLOF['Stellar_Type(1)>MT'][()] >= 7) & (RLOF['Stellar_Type(1)>MT'][()] >= 7) & (RLOF['Stellar_Type(1)>MT'][()] <= 9)
    BinaryStripped_seeds1 = RLOF['SEED'][BinaryStripped_mask1]
    if binaryStripped == True:
        seeds = np.intersect1d(StellarType_seeds1, BinaryStripped_seeds1)
    else:
        seeds = np.setdiff1d(StellarType_seeds1, BinaryStripped_seeds1)
        
    # 3) define the output arrays
    mass1 = [] # SPs['Mass@ZAMS(1)'][StellarType_mask1]
    time1 = []

    # 4) determine time spent stripped
    for seed in seeds:
        sls = printCompasDetails(SLs, seed)
        sls_data = np.array(sls.to_numpy())
        sls_data = sls_data[:, 1:].astype(float)
        primary_mask = sls_data[0][()] == 1
        stripped_mask = sls_data[2][primary_mask] >= 7
        
        t = sls_data[3][primary_mask][stripped_mask][-1]-sls_data[3][primary_mask][stripped_mask][0]
        time1.append(t)

        sps = printCompasDetails(SPs, seed)
        sps_data = np.array(sps.to_numpy())
        sps_data = sps_data[:, 1:].astype(float)

        m = sps_data[6]
        mass1.append(m)

    ## For the secondary companion:
    
    # 1) mask for stellar type, bound vs unbound
    StellarType_mask2 = (SPs['Stellar_Type(2)'][()] == CoType) & (SPs['Unbound'][()] == BoundUnbound)
    StellarType_seeds2 = SPs['SEED'][StellarType_mask2]

    # 2) mask for binary stripped stars
    BinaryStripped_mask2 = (RLOF['Stellar_Type(2)>MT'][()] >= 7) & (RLOF['Stellar_Type(2)>MT'][()] <= 9)
    BinaryStripped_seeds2 = RLOF['SEED'][BinaryStripped_mask2]
    if binaryStripped == True:
        seeds = np.intersect1d(StellarType_seeds2, BinaryStripped_seeds2)
    else:
        seeds = np.setdiff1d(StellarType_seeds2, BinaryStripped_seeds2)
    
    # 3) define the output arrays
    mass2 = [] # SPs['Mass@ZAMS(2)'][StellarType_mask2]
    time2 = []

    # 4) determine time spent stripped
    for seed in seeds:
        sls = printCompasDetails(SLs, seed)
        sls_data = np.array(sls.to_numpy())
        sls_data = sls_data[:, 1:].astype(float)
        secondary_mask = sls_data[0][()] == 2
        stripped_mask = sls_data[2][secondary_mask] >= 7
        
        t = sls_data[3][secondary_mask][stripped_mask][-1]-sls_data[3][secondary_mask][stripped_mask][0]
        time2.append(t)

        sps = printCompasDetails(SPs, seed)
        sps_data = np.array(sps.to_numpy())
        sps_data = sps_data[:, 1:].astype(float)

        m = sps_data[7]
        mass2.append(m)
    
    return [mass1, time1, mass2, time2]
```

```python
bound_HeGB = MassVsTimeSpentStripped(9, 0, True)
bound_HeWD = MassVsTimeSpentStripped(10, 0, True)
bound_COWD = MassVsTimeSpentStripped(11, 0, True)
bound_ONeWD = MassVsTimeSpentStripped(12, 0, True)
bound_NS = MassVsTimeSpentStripped(13, 0, True)
bound_BH = MassVsTimeSpentStripped(14, 0, True)

unbound_HeGB = MassVsTimeSpentStripped(9, 1, True)
unbound_HeWD = MassVsTimeSpentStripped(10, 1, True)
unbound_COWD = MassVsTimeSpentStripped(11, 1, True)
unbound_ONeWD = MassVsTimeSpentStripped(12, 1, True)
unbound_NS = MassVsTimeSpentStripped(13, 1, True)
unbound_BH = MassVsTimeSpentStripped(14, 1, True)
```
